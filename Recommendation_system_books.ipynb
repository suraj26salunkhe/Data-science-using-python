```python
books=data2.pivot_table(index='Userid' , columns='Book_title' , values='Book_rating')
books.head()
```

```python
books.index = data2.Userid.unique()
```

```python
books.fillna(0, inplace=True)
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(books.values,metric='cosine')
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df.index = data2.Userid.unique()
user_sim_df.columns = data2.Userid.unique()
```

```python
def recommend(custId):
    simId = user_sim_df.loc[custId].idxmax()
    simId_book = data2[data2["Userid"]==simId].Book_title
    custId_book=data2[data2["Userid"]==custId].Book_title
    return set (simId_book) - set(custId_book)
```
