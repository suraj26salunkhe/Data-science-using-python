```python
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
data=pd.read_csv("C:/Users/suraj/Desktop/ExcelR/Assignment 9 association rule/my_movies.csv")
data
```

```python
movies=data.iloc[:,5:]
movies
```

```python
#doing with 10% support and 70% confidence
frequent_itemsets = apriori(movies, min_support=0.1, use_colnames=True)

frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
rules
```

```python
rules.sort_values('confidence',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```

```python
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

```python
#doing with 20% support and 80% confidence
frequent_itemsets2=apriori(movies,min_support=0.2,use_colnames=True)
frequent_itemsets2
```

```python
rules2=association_rules(frequent_itemsets2,metric='lift',min_threshold=0.8)
rules2
```

```python
rules2.sort_values('lift',ascending = False)
```

```python
rules2[rules2.lift>1]
```
